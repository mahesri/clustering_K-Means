```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv', index_col=0)
data.head()
```

```python
data['cluster'].value_counts()
```

```python
x_ = data['x']
y_ = data['y']
label = data['cluster']
plt.scatter(x_, y_, c = label, cmap='plasma')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
x = data[['x', 'y']]
y = data['cluster']
x.head()
```

```python
#praproses
scaler = StandardScaler()
x = scaler.fit_transform(x)
x[:5]
```

```python
kmeans = KMeans(n_clusters=3, random_state=30)
kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
x_ = data['x']
y_ = data['y']
label = kmeans.labels_
plt.scatter(x_, y_, c = label, cmap='plasma')
```

```python
#ini bagian evaluasi
#menggunakan square errors

sse = []
index = range(1, 10)
for i in index:
  kmeans = KMeans(n_clusters=i, random_state=30)
  kmeans.fit(x)
  sse_ = kmeans.inertia_
  sse.append(sse_)
  print(i,sse_)
```

```python
plt.plot(index, sse)
plt.xlabel('n_cluster')
plt.ylabel('SSE')
plt.show()
```

```python
#EVAL disini pakai ran score
from sklearn import metrics
rand = []
index = range(1, 10)
for i in index:
  kmeans = KMeans(n_clusters=i, random_state=30)
  kmeans.fit(x)
  rand_ = metrics.adjusted_rand_score(y, kmeans.labels_)
  rand.append(sse_)
  print(i, sse_)

```

```python
plt.plot(index, sse)
plt.xlabel('n_cluster')
plt.ylabel('rand score')
plt.show()
```
